```python
# Read parameters injected by the bundle YAML
dbutils.widgets.text('catalog', 'workspace')
dbutils.widgets.text('schema', 'dev')

catalog = dbutils.widgets.get('catalog')
schema = dbutils.widgets.get('schema')

print(f'Target catalog: {catalog}')
print(f'Target schema:  {schema}')
```

```python
# Ensure target schema exists
spark.sql(f'CREATE SCHEMA IF NOT EXISTS {catalog}.{schema}')
print(f'Schema {catalog}.{schema} ready')
```

```python
# Bronze ingest: copy samples.tpch.customer to our bronze layer
source_table = 'samples.tpch.customer'
target_table = f'{catalog}.{schema}.bronze_customer'

df = spark.read.table(source_table)
row_count = df.count()
print(f'Source rows: {row_count}')

df.write.mode('overwrite').saveAsTable(target_table)
print(f'Wrote to {target_table}')
```

```python
# Verify
result = spark.sql(f'SELECT COUNT(*) as row_count FROM {target_table}').collect()[0]
print(f'Bronze table row count: {result["row_count"]}')
```
